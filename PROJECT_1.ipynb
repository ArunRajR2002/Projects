## 4.What is the average Monthly Loan Payment across different Loan Duration categories?

```python
avg_monthly_payment = df.groupby('Loan_Duration')['Monthly_Loan_Payment'].mean()
avg_monthly_payment.plot(kind='bar', color=sns.color_palette("magma"))
plt.title('Average Monthly Loan Payment by Loan Duration')
plt.xlabel('Loan Duration')
plt.xticks(rotation=45)
plt.ylabel('Average Monthly Loan Payment')
plt.show()
```

### The analysis shows that the average monthly loan payment decreases as the loan duration increases. Borrowers opting for shorter loan durations face higher monthly payments, likely due to the need to repay the loan within a shorter period. Conversely, those who choose longer loan terms have lower monthly payments, making the loans more manageable in the short term, although they may pay more in total over the life of the loan due to accumulated interest.

## 6.IS THERE A RELATIONSHIP BETWEEN MONTHLY INCOME AND DEBT TO INCOME RATIO?

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

bins = [0, 2000, 4000, 6000, 8000, 10000, np.inf]
labels = ['<2000', '2000-4000', '4000-6000', '6000-8000', '8000-10000', '10000+']
df['IncomeBin'] = pd.cut(df['Monthly_Income'], bins=bins, labels=labels, right=False)

avg_dti_by_income = df.groupby('IncomeBin')['Debt_To_Income_Ratio'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(data=avg_dti_by_income, x='IncomeBin', y='Debt_To_Income_Ratio', marker='o')
plt.title('Average Debt-to-Income Ratio by Monthly Income')
plt.xlabel('Monthly Income Range')
plt.ylabel('Average Debt-to-Income Ratio')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```


### THE ANALYSIS OF RELATIONSHIP BETWEEN MONTHLY INCOME AND DEBT TO INCOME RATIO REVEALS DISTINCT TRENDS.IT APPEARS THAT AS THE MONTHLY INCOME INCREASES THE AVERAAGE DTI INCREASES UPTO MONTHLY INCOME 6,000-8,000, THEN IT DECREASES AS THE ,MONTLY INCOME INCREASES

## 7.HOW DOES EMPLOYMENT STATUS INFLUENCE THE RISK SCORE?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

avg_risk_by_employment = df.groupby('Employment_Status')['Risk_Score'].mean().reset_index()
avg_risk_by_employment = avg_risk_by_employment.sort_values(by='Employment_Status')

plt.figure(figsize=(10, 6))
sns.lineplot(data=avg_risk_by_employment, x='Employment_Status', y='Risk_Score', marker='o')
plt.title('Average Risk Score by Employment Status')
plt.xlabel('Employment Status')
plt.ylabel('Average Risk Score')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```


### THE ANALYSIS OF AVERAGE RISK SCORE ACROSS DIFFERENT EMPLOYMENT STATUS REVEALS SIGNIFICANT INSIGHTS INTO  HOW EMPLOYMENT INFLUENCES FINANCIAL RISK ASSESMENTS.THE LINE PLOT INDICATES THAT THOSE WHO ARE EMPLOYED HAVE LOWER AVERAGE RISK SCORE,SUGGESTING THAT STABLE EMPLOYMENT IS ASSOCIATED WITH GREATER FINANCIAL RELIABILITY.CONVERSELY,THOSE WHO ARE SELF-EMPLOYED OR UNEMPLOYED TENDS TO HAVE HIGHER AVERAGE RISK SCORE

## 9.WHAT IS THE CORRELATION BETWEEN DEBT-TO-INCOME RATIO AND LOAN AMOUNT?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
correlation = df['Debt_To_Income_Ratio'].corr(df['Loan_Amount'])
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Debt_To_Income_Ratio', y='Loan_Amount', data=df)
plt.title('Debt-to-Income Ratio vs Loan Amount')
plt.xlabel('Debt_to_Income Ratio')
plt.ylabel('Loan Amount')
plt.axhline(y=df['Loan_Amount'].mean(), color='r',linestyle='-', label='Mean Loan Amount')
plt.axvline(x=df['Debt_To_Income_Ratio'].mean(), color='g',linestyle='-',label='Mean Debt-to-Income Ratio')
plt.legend()
plt.show()
```


### THE ANALYSIS REVEALS A CORRELATION COEFFICIENT OF APPROXIMATELY r=0.25.THIS VALUE REVEALS A MODERATE POSITIVE CORRELATION BETWEEN DEBT-TO-INCOME RATIO AND LOAN AMOUNT.

## 10. HOW DOES RISK SCORE VARY ACROSS DIFFERENT LEVELS OF ANNUALINCOME,LOANAMOUNT AND CREDITSCORE?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.pairplot(df, vars=['Risk_Score', 'Annual_Income', 'Loan_Amount', 'Credit_Score'], diag_kind='kde', kind='scatter')
plt.suptitle('Pairplot of RiskScore, AnnualIncome, LoanAmount, and CreditScore', y=1.02, fontsize=15)
plt.show()
```

  ### RiskScore has no strong linear relationship with either AnnualIncome or LoanAmount, but there is a slight positive association with CreditScore.
  ### AnnualIncome and LoanAmount show a moderate positive relationship, meaning higher-income individuals tend to request larger loans.
  ### CreditScore does not have a strong connection to either AnnualIncome or LoanAmount, but it does show some degree of positive relationship with RiskScore.
## Overall, the pairplot reveals that credit score may be a slightly better indicator of RiskScore than income or loan amount, but these relationships are not very strong.
